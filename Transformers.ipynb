```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
classifier(["I love you","I hate you"])
```

```python
classifier("I was not happy in past but now some what good")
```

```python
#zero-shot classification pipeline helps us to select the labels for classification
classifier = pipeline("zero-shot-classification")
classifier(
           "This is a course on transformers library",
           candidate_labels=["education","business","politics"],
           )
```

```python
#Text-generation pipeline uses an input prompt to generate text
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator("In this course, we will teach you how to", max_length=30, num_return_sequences=2)
```

```python
#Fill mask pipeline will predict missing word in sentences
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models", top_k=2)
```

```python
#NER pipeline identifies entities such as persons,organization or locations in a sentences
ner = pipeline("ner", grouped_entities=True)
ner("My name is Balaji and I am a Data Scientist in Chennai")
```

```python
#Question answering pipeline extracts answers to a question from  given text
question_answering = pipeline("question-answering")
question_answering(
    question="Where do I work?",
    context="My name is Balaji and I am a Data Scientist in Chennai",
)
```

```python
#Summarizer pipeline creates summarises the given text
summarizer = pipeline("summarization")
summarizer("""
Long Short Term Memorey(LSTM). Its an artificial recurrent neural network(LSTM-RNN) used in the field of deep learning and NLP.
It was developed at 1977 by Sepp Hochreiter and Jurgen Schmid Huber. Unlike standard feed forward neural network, LSTM have feedback connections i,e it can process not only single data point but also entire sequence of data.
LSTM are modified version of RNN which makes it easier to remember past data in memorey. LSTM are explicitly designed to avoid long term dependency problem, remembering information for a long period of time is practically their default behavior's.
LSTM also have chain like structure but the repeating module has different structure like forget gate, input gate, output gate and cell state
""")
```

```python
#Translator pipeline translates text from one lang to another
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("This is a course on transformers library")
```

```python
#Auto-Tokenizer class can load the tokenizer  for any checkpoints -- Tokenizer
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "This course is amazing!",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
inputs
#tokenizer("This is a course on transformers library")
```

```python
#TFAutoModel class loads a model witjout any pretrained head --Model
from transformers import TFAutoModel
model = TFAutoModel.from_pretrained("distilbert-base-uncased")
import tensorflow as tf
inputs = {
    'input_ids': tf.convert_to_tensor(inputs['input_ids']),
    'attention_mask': tf.convert_to_tensor(inputs['attention_mask'])
}
outputs = model(inputs)
outputs.last_hidden_state.shape
```

TensorShape([2->batch_size,16->sequence_length,768->hidden_state])

```python
#For Each TFAutoModelForXX loads a model suitable for specific task
from transformers import TFAutoModelForSequenceClassification
model = TFAutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
outputs = model(inputs)
outputs.logits
```

```python
#to go for logic propabilites we apply softmax layer  --PostProcessing
import tensorflow as tf
predictions = tf.nn.softmax(outputs.logits, axis=-1)
predictions
```

```python
#The model configuration contains the index to label mapping
model.config.id2label
```

The above array says
first sentence : Negative 4.02% Positive 95.98%
second sentence : Positive 99.46% Negative 0.54%

```python
#TFAutoModel API allows us to instantiate a pretrained model from any checkpoints
from transformers import TFAutoModel
Bert_model = TFAutoModel.from_pretrained("distilbert-base-uncased")
print(type(Bert_model))
GPT_model = TFAutoModel.from_pretrained("gpt2")
print(type(GPT_model))
BART_model = TFAutoModel.from_pretrained("facebook/bart-large-cnn")
print(type(BART_model))
```
