```python
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline
```

```python
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10, 5
```

```python
zero_shot_classification = pipeline("zero-shot-classification")
```

```python
result = zero_shot_classification(
    sequences = "Can you order some Pizza?",
    candidate_labels = ["Flight Travel","Cabs Travel","Food","Movies"],
    multi_class = True
)
```

```python
plt.bar(result["labels"],result["scores"])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```

```python
result = zero_shot_classification(
    sequences = "Can you order some Pizza & book an Uber to the nearest cinema House at 10PM?",
    candidate_labels = ["Flight Travel","Cabs Travel","Food","Movies"],
    multi_class = True
)
```

```python
plt.bar(result["labels"],result["scores"])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```

```python
result = zero_shot_classification(
    sequences = "Can you order some Pizza & book an Uber once my Flight reaches Mumbai to the nearest cinema House at 10PM?",
    candidate_labels = ["Flight Travel","Cabs Travel","Food","Movies"],
    multi_class = True
)
```

```python
plt.bar(result["labels"],result["scores"])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```
